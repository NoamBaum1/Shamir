## How does it work ?


Share refreshing function use a polynomial characteristic to work
As a polynom f(x) is written like :

$$f(x) = a + bx + cx² + dx³ ...$$

Adding two polynomials concludes in adding there factors : 

$f(x) = a_f + b_fx + c_fx²  ...$
$h(x) = a_h + b_hx + c_hx²  ...$
$\iff f(x) + h(x) = (a_f +a_h)+ (b_f + b_h)x +( c_f + c_h)x² ...$

As $f(0) = a , a$ is the secret.

But if $h(0) = 0$
$\iff f(0) + h(0) = a$

With this characteristic some polynomials can be added to the original encoding polynomial without changing the secret
